# **Notebook em linguagem R**

Conceitos básicos em linguagem R para a disciplina de Estatística.

```r
# Coletar Dados
venda <- 1.59
lucro <- 40

# Processar dados
lucro_per <- lucro/100

custo <- round(
    venda / (1 + lucro_per),
    2
)

# imprimir resultado
print(paste('O preço de custo é: R$', custo))

```

## **Vetores**

Os vetores em R são equivalentes ao conceito de lista em Python.

```r
# Coletar valor das notas
n1 <- 6.5
n2 <- 4.2
n3 <- 8.9

# Armazenar as três notas em um conjunto (vetor)
notas <- c(n1, n2, n3)

# Calcular a média das notas a partir do vetor
media <- round(
  mean(notas),
  1)

# Contar quantas notas há no vetor
qtde_notas <- length(notas)
print(paste("Quantidade de notas:", qtde_notas))

# Apresentar o resultado
print(paste("Média:", media))
```
