# **Medidas de Dispersão**

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
```

```python
# Carregar os dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmA.xlsx",
                   sheet_name="DadosTratados2")

df.info()
```

```python
# Selecionar variáveis
idades_outras = df[df['Cidade_Origem'] != 'Blumenau']['Idade'].tolist()

idades_bnu = df[df['Cidade_Origem'] == 'Blumenau']['Idade'].tolist()

print(idades_outras)
print(idades_bnu)
```

```python
########################################
#### Calculo de medidas de posição #####
########################################

# Amplitude total
amp_outras = np.max(idades_outras)-np.min(idades_outras)
amp_bnu = np.nanmax(idades_bnu)-np.nanmin(idades_bnu)

# Desvio médio
mae_outras = np.mean(np.abs(idades_outras - np.mean(idades_outras)))
mae_bnu = np.nanmean(np.abs(idades_bnu - np.nanmean(idades_bnu)))

# Variância
var_outras = np.var(idades_outras, ddof=1)
var_bnu = np.nanvar(idades_bnu, ddof=1)

# Desvio padrão
std_outras = np.std(idades_outras, ddof=1)
std_bnu = np.nanstd(idades_bnu, ddof=1)

# Coeficiente de variação
cv_outras = (std_outras/np.mean(idades_outras))*100
cv_bnu = (std_bnu/np.nanmean(idades_bnu))*100


# Imprimir resultados
print(f'Média (outras cidades) = {np.mean(idades_outras):.2f}')
print(f'Média (Blumenau) = {np.nanmean(idades_bnu):.2f}\n')
print(f'Amplitude (outras cidades) = {amp_outras}')
print(f'Amplitude (Blumenau) = {amp_bnu}\n')
print(f'Desvio médio (outras cidades) = {mae_outras:.2f}')
print(f'Desvio médio (Blumenau) = {mae_bnu:.2f}\n')
print(f'Variância (outras cidades) = {var_outras:.2f}')
print(f'Variância (Blumenau) = {var_bnu:.2f}\n')
print(f'DesvPad (outras cidades) = {std_outras:.2f}')
print(f'DesvPad (Blumenau) = {std_bnu:.2f}\n')
print(f'CV (outras cidades) = {cv_outras:.1f}')
print(f'CV (Blumenau) = {cv_bnu:.1f}')
```

# **Definição de Funções**

Automatizar a cálculo de estatísticas descritivas de conjunto de dados

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import statistics
```

```python
# Carregar a base de dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmA.xlsx",
                   sheet_name="DadosTratados2")

df.info()
```

```python
# Definir funções para Estatística Descrita

def descritiva_quanti(dataset, variavel):
  """
  Parâmetros de entrada:
  dataset: um dataframe
  variável: nome de um variável "string"
  """
  # 1.Filtrar a variável analisada
  coluna = dataset[variavel]

  # 2.Calcular medidas de posição
  media   = round(np.nanmean(coluna), 1)
  mediana = round(np.nanmedian(coluna), 1)

  # 3.Calcular medidas de dispersão
  min = round(np.min(coluna), 1)
  max = round(np.max(coluna), 1)
  amp = max - min
  var = round(np.nanvar(coluna, ddof=1), 1)
  std = round(np.nanstd(coluna, ddof=1), 1)
  cv  = round((std/media) * 100, 1)

  # 4.Agrupar os resultados em um dicionário
  resultados = {
      'Variavel': variavel,
      'Media': media,
      'Mediana': mediana,
      'Mínimo': min,
      'Max': max,
      'Amplitude': amp,
      'Variância': var,
      'Desvio Padrão': std,
      'CV (%)': cv}

  # 5.Converter o dicionário em um DataFrame
  df_final = pd.DataFrame([resultados])

  # 6.Definir qual será a saída da função
  return df_final


def descritiva_quanli(dataset, variavel):
  """
  Parâmetros de entrada:
  dataset: um dataframe
  variável: nome de um variável "string"
  """
  # 1.Filtrar a variável analisada
  coluna = dataset[variavel]

  # 2.Calcular medidas de posição
  moda = statistics.multimode(coluna)

  # 3.Calcular medidas de dispersão
  ### Razão de Variação
  n   = len(coluna.dropna())
  f_m = coluna.value_counts().max() #frequência da moda
  rv  = round(1 - (f_m/n), 1)

  # 4.Agrupar os resultados em um dicionário
  resultados = {
      'Variavel': variavel,
      'Moda': moda,
      'RV': rv}

  # 5.Converter o dicionário em um DataFrame
  df_final = pd.DataFrame([resultados])

  # 6.Definir qual será a saída da função
  return df_final
```

```python
# Calcular resumos estatísticos de dados quantitativos

resumo_quanti = descritiva_quanti(df ,"Leituras")
resumo_quanti
```

```python
# Calcular resumos estatísticos de dados qualitativos

resumo_quanli = descritiva_quanli(df ,"Cidade_Origem")
resumo_quanli
```
