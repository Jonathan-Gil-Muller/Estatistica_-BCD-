## **Interagindo com Pytyhon: input e Listas**

```python
nome = input("Qual seu nome?")

n1 = float(input("Digite sua nota 1:"))
n2 = float(input("Digite sua nota 2:"))
n3 = float(input("Digite sua nota 3:"))
```

```python
lista_notas = [n1, n2, n3]

soma = sum(lista_notas) #A função 'sum' soma o valor dos objetos que estão dentro da lista
qtade_notas = len(lista_notas)  #A função 'len' conta quantos objetos que estão dentro da lista

media = soma/qtade_notas

print(soma)
print(qtade_notas)
print(f"A média das notas é {media:.2f}")
```

```python
import numpy as np

media_lista = np.mean(lista_notas)
print(f"A média das notas é {media_lista:.2f}")
```

```python
"""
PROBLEMA: Calcular a n3 para aprovação, dados
a nota 1, nota2 e média mínima para aprovação

"""

# COLETAR INFORMAÇÕES INICIAIS
media_apr = float(input("Qual é a média para aprovação?")) # média mínima
nota1 = float(input("Qual foi a nota na prova 1?"))        # nota 1
nota2 = float(input("Qual foi a nota na prova 2?"))        # nota 2

# PROCESSAMENTO DOS DADOS
nota3 = media_apr * 3 - nota1 - nota2

# IMPRIMIR O RESULTADO
print("A nota necessária na avaliação 3 é:", round(nota3, 2))
print(f"A nota necessária na avaliação 3 é: {nota3:.1f}")
```

**PROBLEMA 2**: Informar o preço de custo para um usuário conhecendo o preço de venda e a margem de lucro.

```python
# 1º PASSO: ENTRADA DOS DADOS

preco_venda = 2.59
margem_lucro = 35 #em valor percentual

# 2º PASSO: PROCESSMAENTO DOS DADOS

preco_custo = preco_venda / (1 + margem_lucro/100)

# 3º PASSO: IMPRIME OS RESULTADOS

print(f"O preço de custo é: R${round(preco_custo, 2)}")
```

## **Dicionários e DataFrame**

```python
# Armazenar vários conjuntos de dados em um único objeto, o dicionário
dicionario = {
    'Nome': ["Aluno 1", "Aluno 2", "Aluno 3"],
    'Nota 1': [4.5, 9.8, 7.1],
    'Nota 2': [5.0, 8.2, 3.8],
    'Nota 3': [6.5, 8.3, 9.1]}

print(dicionario)
```

```python
# Importar uma biblioteca para trabalhar com DataFrames
import pandas as pd

# Converter o dicionário em um DataFrame

df = pd.DataFrame(dicionario)

df.head()
#print(df)
```

```python
# Inserir uma nova coluna no df com o valor das médias

df["Médias"] = (df["Nota 1"] + df["Nota 2"] + df["Nota 3"]) / 3
df["Médias"] = round(df["Médias"], 1)

df.head()
```

```python
import numpy as np

media_n3 = np.mean(df['Nota 3'])

print(f"A médias das Notas 3 é: {media_n3}")
#print(f"A tabela é: \n\n {df}")
```
