## **Variáveis**

```python
# Definir as variáveis
disciplina   = 'Estatística' #variável string (str)
qtade_alunos = 49            #variável interira (int)
aprovados    = 30            #variável interira (int)

# Cálculo da taxa de aprovação
"""A função round é utilizada para fazer arredondamento estatística
A multiplicação por 100 foi realizada para converter o valor em
percentual"""
taxa_aprovacao = aprovados/qtade_alunos
taxa_aprovacao_arr = round((aprovados/qtade_alunos)*100 , 2)

#Imprimir o resultado
print(f"A taxa de aprovação é: {taxa_aprovacao_arr}%")
#ou
print(f"A taxa de aprovação é: {taxa_aprovacao:.1%}")
```

## **Interagindo com o Python**
Função 'input'

```python
# Entrada dos dados
nome = input('Qual é o seu nome? ')
n1 = float(input('Digite a primeira nota: '))
n2 = float(input('Digite a segunda nota: '))
n3 = float(input('Digite a terceira nota: '))

# Processamento dos dados
media = round(((n1 + n2 + n3)/3), 1)

# Imprimir o resultado
print(f'A média do aluno {nome} é: {media}')
```

## **Listas**

```python
# Armazenar as três notas em uma lista
lista_notas = [n1, n2, n3]

# Calcular a média das notas a partir da lista
media_lista = sum(lista_notas) / len(lista_notas)

# Imprimir o resultado
print(f'A média do aluno {nome} é: {media_lista}')
```

### *Média de lista com uso de biblioteca*

```python
# Importar a biblioteca
import numpy as np

# Calcular a média da lista utilizando a função 'mean' do numpy
media_lista_bibl = np.mean(lista_notas)

# Imprimir resultado
print(f'A média do aluno {nome} é: {media_lista_bibl}')
```

**Problema**: desenvolver um algoritmo que o usuário informe a média mínima de aprovação, as duas notas que ela já possui na disciplina e o algoritmo retorne a nota mínima que o estudante precisa atingir para conseguir aprovação.
image.png

```python
# Coletar informação
nota_min = float(input('Qual a nota mínima para aprovação? '))
N1 = float(input('Nota 1: '))
N2 = float(input('Nota 2: '))

# Processar as informações
N3 = round(nota_min*3 - N1 - N2 , 1)

# Imprimir o resultado
print(f'Para aprovação você precisa de, no mínimo, {N3} na avaliação 3.')
```
