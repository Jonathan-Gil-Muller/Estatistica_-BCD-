## **Medidas de dispersão**

```python
# Importar as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Carrergar os dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmB.xlsx",
                   sheet_name="DadosTratados2")

df.info()
```

```python
# Selecionar as variáveis que serão analisadas
leitura_bnu = df[df['C_origem'] == 'Blumenau']['Leituras'].tolist()
print(leitura_bnu)

leitura_outros = df[df['C_origem'] != 'Blumenau']['Leituras'].tolist()
print(leitura_outros)
```

```python
# Medidas de Posição
media_bnu    = np.mean(leitura_bnu)
media_outros = np.mean(leitura_outros)

mediana_bnu    = np.median(leitura_bnu)
mediana_outros = np.median(leitura_outros)

# Calcular medidas de dispersão

### Amplitude total
amp_bnu = max(leitura_bnu) - min(leitura_bnu)
amp_out = max(leitura_outros) - min(leitura_outros)

### Desvio Médio
mae_bnu = np.mean(np.abs(leitura_bnu - media_bnu))
mae_out = np.mean(np.abs(leitura_outros - media_outros))

### Variância
var_bnu = np.var(leitura_bnu, ddof=1)
var_out = np.var(leitura_outros, ddof=1)

### Desvio Padrão
std_bnu = np.std(leitura_bnu, ddof=1)
std_out = np.std(leitura_outros, ddof=1)

### Coeficiente de Variação
cv_bnu = (std_bnu / media_bnu)*100
cv_out = (std_out / media_outros)*100

# Imprimir resultados
print(f'Blumenau: Média = {media_bnu:.2f}, Mediana = {mediana_bnu:.2f}')
print(f'Outros: Média = {media_outros:.2f}, Mediana = {mediana_outros:.2f}\n')
print(f'Blumenau: Amplitude = {amp_bnu:.2f}')
print(f'Outros: Amplitude = {amp_out:.2f}\n')
print(f'Blumenau: MAE = {mae_bnu:.2f}')
print(f'Outros: MAE = {mae_out:.2f}\n')
print(f'Blumenau: Variância = {var_bnu:.2f}')
print(f'Outros: Variância = {var_out:.2f}\n')
print(f'Blumenau: Desvio Padrão = {std_bnu:.2f}')
print(f'Outros: Desvio Padrão = {std_out:.2f}\n')
print(f'Blumenau: Coeficiente de Variação = {cv_bnu:.2f}%')
print(f'Outros: Coeficiente de Variação = {cv_out:.2f}%')
```

# **Definição de Funções**

```python
# Importar as bibliotecas
import pandas as pd
import numpy as np
import statistics
```

```python
# Carrergar os dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmB.xlsx",
                   sheet_name="DadosTratados2")
```

```python
def descritiva_quanti(df, variavel):

  # df = DataFrame com dados
  # variável = Nome da coluna analisada

  #1. Filtrar a variável do DataFrame
  coluna = df[variavel]

  #2. Calcular as medidas de posição
  media = round(np.mean(coluna), 2)
  mediana = round(np.median(coluna), 2)

  #3. Calcular as medidas de dispersão
  maximo = round(max(coluna), 2)
  minimo = round(min(coluna), 2)
  amplitude = round(maximo - minimo, 2)
  desvpad = round(np.std(coluna, ddof=1), 2)
  cv = round((desvpad / media)*100, 2)

  #4. Criar um dicionário com os resultados
  resultados = {
      "Variável":variavel,
      'Média': media,
      'Mediana': mediana,
      'Máximo': maximo,
      'Mínimo': minimo,
      'Amplitude': amplitude,
      'Desvio Padrão': desvpad,
      'Coeficiente de Variação': cv}

  #5. Converter o diconário dos resultados em um DataFrame
  df_final = pd.DataFrame([resultados])

  #6. Definir o retorno (saída) da função
  return df_final
```

```python
def descritiva_quali(df, variavel):

  # df = DataFrame com dados
  # variável = Nome da coluna analisada (string)

  #1. Filtrar a variável do DataFrame
  coluna = df[variavel]

  #2. Calcular as medidas de posição
  moda = statistics.multimode(coluna)

  #3. Calcular as medidas de dispersão
  n   = len(coluna.dropna())        # Total de Valores não nulos
  f_m = coluna.value_counts().max() # Frequência Máxima
  # Cálculo da razão de Variância
  rv = round(1 - (f_m/n),2)

  #4. Criar um dicionário com os resultados
  resultados = {
      "Variável":variavel,
      'Moda': moda,
      'RV': rv}

  #5. Converter o diconário dos resultados em um DataFrame
  df_final = pd.DataFrame([resultados])

  #6. Definir o retorno (saída) da função
  return df_final
```

# **Cálculo do Tamanho da amostra**

```python
import numpy as np
```

## _População **Infinita**_

### Dados **quanti**tativos

```python
def amostra_quantitativa_infinita(Z, desvio_padrao, E):
    """
    Calcula o tamanho da amostra para dados quantitativos em populações infinitas.

    Parâmetros:
    Z (float): Valor crítico da distribuição normal (ex: 1.96 para 95% de confiança).
    desvio_padrao (float): Desvio padrão estimado ou obtido via piloto (sigma).
    E (float): Margem de erro admissível na mesma unidade de medida dos dados.

    Retorna:
    (float): Tamanho amostral arredondado para cima.
    """
    # Elevando o desvio padrão ao quadrado para obter a variância (sigma^2)
    variancia = desvio_padrao ** 2

    # Aplicando a estrutura da fórmula: (Z^2 * sigma^2) / E^2
    n = ((Z ** 2) * variancia) / (E ** 2)

    # Retorna o inteiro imediatamente superior
    return np.ceil(n)
```

```python
# TESTE DA FUNÇÃO

# Definir as variáveis
fator_z = 1.96
sigma = 4
erro = 0.5

# Execução do cálculo
tamanho_n = amostra_quantitativa_infinita(
    Z=fator_z,
    desvio_padrao=sigma,
    E=erro)

print(f"O tamanho amostral mínimo calculado é: {tamanho_n} indivíduos.")
```

### Dados **quali**tativos


```python
def amostra_qualitativa_infinita(Z, p, E):
    """
    Calcula o tamanho da amostra para dados qualitativos (proporções) em populações infinitas.

    Parâmetros:
    Z (float): Valor crítico da distribuição normal (ex: 1.96 para 95% de confiança).
    p (float): Proporção estimada do evento de interesse na população (ex: 0.50 para 50%).
    E (float): Margem de erro admissível expressa em formato decimal (ex: 0.05 para 5%).

    Retorna:
    float: Tamanho amostral mínimo (n) arredondado para cima.
    """
    # Calculando a proporção complementar (fracasso)
    q = 1 - p

    # Aplicando a estrutura da fórmula: (Z^2 * p * q) / E^2
    n = ((Z ** 2) * p * q) / (E ** 2)

    # np.ceil(n) arredonda para cima; .astype(int) converte o float do numpy para inteiro puro
    return np.ceil(n)
```

```python
# TESTE DA FUNÇÃO

# Definindo os parâmetros:
fator_z = 1.96
proporcao_p = 0.50
erro_toleravel = 0.05

# Execução do cálculo
tamanho_n = amostra_qualitativa_infinita(
    Z=fator_z,
    p=proporcao_p,
    E=erro_toleravel)

print(f"O tamanho amostral mínimo calculado é: {tamanho_n} indivíduos.")
```

## _População **Finita**_

### Dados **quanti**tativos

```python
def amostra_quantitativa_finita(N, Z, E, sigma):
    """
    Calcula o tamanho da amostra para dados quantitativos em populações finitas.

    Parâmetros:
    Z (float): Valor crítico da distribuição normal.
    sigma (float): Desvio padrão estimado ou obtido via piloto (sigma).
    E (float): Margem de erro admissível na mesma unidade de medida dos dados.
    N (int): Tamanho total da população.

    Retorna:
    float: Tamanho amostral mínimo (n) arredondado para cima.
    """
    # Elevando o desvio padrão ao quadrado para obter a variância (sigma^2)
    variancia = sigma ** 2

    # Calculando o numerador: Z^2 * sigma^2 * N
    numerador = (Z ** 2) * variancia * N

    # Calculando o denominador: E^2 * (N - 1) + Z^2 * sigma^2
    denominador = (E ** 2) * (N - 1) + (Z ** 2) * variancia

    # Aplicando a estrutura da fórmula
    n = numerador / denominador

    # np.ceil(n) arredonda para cima; .astype(int) converte o float do numpy para inteiro
    return np.ceil(n)
```

```python
# TESTE DA FUNÇÃO

# Definir as variáveis
N = 13500
fator_z = 1.96
sigma = 4
erro = 0.5

# Execução do cálculo
tamanho_n = amostra_quantitativa_finita(N, fator_z, erro, sigma)

print(f"O tamanho amostral mínimo calculado é: {tamanho_n} indivíduos.")
```

### Dados **quali**tativos

```python
def amostra_qualitativa_finita(Z, p, E, N):
    """
    Calcula o tamanho da amostra para dados qualitativos (proporções) em populações finitas.

    Parâmetros:
    Z (float): Valor crítico da distribuição normal.
    p (float): Proporção estimada do evento de interesse na população.
    E (float): Margem de erro percentual admissível expressa em formato decimal.
    N (int): Tamanho total da população.

    Retorna:
    float: Tamanho amostral mínimo (n) arredondado para cima.
    """
    # Calculando a proporção complementar (fracasso)
    q = 1 - p

    # Calculando o numerador: Z^2 * p * q * N
    numerador = (Z ** 2) * p * q * N

    # Calculando o denominador: E^2 * (N - 1) + Z^2 * p * q
    denominador = (E ** 2) * (N - 1) + (Z ** 2) * p * q

    # Aplicando a estrutura da fórmula
    n = numerador / denominador

    # np.ceil(n) arredonda para cima; .astype(int) converte o float do numpy para inteiro
    return np.ceil(n)
```

```python
# ==============================================================================
# MARGEM DE ERRO FINAL (QUANTITATIVO)
# ==============================================================================

# 1. Encontrar o valor correspondente de Z para o nível de confiança estipulado
z = stats.norm.ppf((1 + confianca) / 2)

# 2. Calcular a margem de erro final (ou real)
margem_erro = z * erro_padrao_1

print(f"Valor de Z: {z:.2f}")
print(f"Estimativa pontual (média amostral):   {dados_amostra.mean():.2f}")
print(f"Margem de Erro:   {margem_erro:.2f}")
```

### Função para análise estatística completa de dados amostrais **QUANTI**TATIVOS

```python
def analisar_dados_amostrais_quantitativos(dados, confianca=0.95, unidade='minutos'):
    """
    Realiza a análise estatística completa de dados amostrais quantitativos.
    Calcula média, desvio padrão, erro padrão, margem de erro, intervalo de confiança
    e gera um relatório final da pesquisa.

    Parâmetros:
    dados (list ou np.array): Lista ou array com os dados quantitativos coletados.
    confianca (float): Nível de confiança desejado (padrão: 0.95 para 95%).
    unidade (str): String com a unidade de medida dos dados da amostra.
    """
    # Converter a entrada para um array do numpy para garantir os cálculos vetoriais
    dados_amostra = np.array(dados)

    # 1. Tamanho da amostra (n) e Estimativa Pontual (Média)
    n = len(dados_amostra)
    media = np.mean(dados_amostra)

    # 2. Desvio Padrão Amostral (s)
    desvio_padrao = np.std(dados_amostra, ddof=1)

    # 3. Erro Padrão da Média (EP)
    erro_padrao = stats.sem(dados_amostra)

    # 4. Encontrar o valor crítico Z
    z = stats.norm.ppf((1 + confianca) / 2)

    # 4. Margem de Erro Real (E_real) - Z * EP
    margem_erro = z * erro_padrao

    # 5. Intervalo de Confiança (Limites Inferior e Superior)
    limite_inferior = media - margem_erro
    limite_superior = media + margem_erro

    # 6. EXIBIÇÃO DO PAINEL DE MÉTRICAS
    print("-" * 60)
    print(f"PAINEL DE MÉTRICAS DA ANÁLISE (Confiança: {confianca*100:.0f}%)")
    print("-" * 60)
    print(f"Tamanho da Amostra (n):        {n} respostas")
    print(f"Média Amostral (Estimativa):   {media:.1f} {unidade}")
    print(f"Desvio Padrão Amostral (s):    {desvio_padrao:.1f} {unidade}")
    print(f"Erro Padrão da Média (EP):     {erro_padrao:.1f} {unidade}")
    print(f"Margem de Erro Final (E_real): {margem_erro:.1f} {unidade}")
    print(f"Intervalo de Confiança (IC):   [{limite_inferior:.0f} {unidade} até {limite_superior:.0f} {unidade}]")
    print("-" * 60)

    # ==============================================================================
    # GERAÇÃO DO TEXTO DO RELATÓRIO AUTOMATIZADO
    # ==============================================================================
    print("\nRELATÓRIO TÉCNICO OFICIAL:")
    print("=" * 60)
    print(
    f"Com base na coleta de dados realizada junto a uma amostra probabilística\n"
    f"com nível de confiança em {confianca*100:.0f}%,afirma-se que a média populacional \n"
    f"do parâmetro pesquisado encontra-se no intervalo compreendido entre \n"
    f"{limite_inferior:.1f} e {limite_superior:.1f} {unidade}."
)
    print("=" * 60)
```

### Função para análise estatística completa de dados amostrais **QUALI**TATIVOS

```python
def analisar_dados_amostrais_qualitativos(dados, valor_alvo, confianca=0.95):
    """
    Realiza a análise estatística de dados amostrais qualitativos (proporções)
    para um candidato específico em cenários de múltiplos candidatos.

    Parâmetros:
    dados (list): Lista com as respostas textuais coletadas em campo.
    valor_alvo (str): Nome do candidato que deseja analisar (ex: "A").
    confianca (float): Nível de confiança desejado (padrão: 0.95).
    """
    # 1. Contagem total da amostra (n)
    n = len(dados)

    # 2. Contar quantos votos o candidato alvo teve e calcular a proporção real (p)
    votos_candidato = dados.count(valor_alvo)
    p = votos_candidato / n
    q = 1 - p  # Proporção de todos os outros candidatos combinados

    # 3. Calcular o Erro Padrão (EP) para Proporções
    erro_padrao = np.sqrt((p * q) / n)

    # 4. Encontrar o valor crítico Z
    z_critico = stats.norm.ppf((1 + confianca) / 2)

    # 5. Calcular a Margem de Erro Final (E_real) - Z * EP
    margem_erro = z_critico * erro_padrao

    # 6. Calcular o Intervalo de Confiança (Convertendo para porcentagem)
    limite_inferior = (p - margem_erro) * 100
    limite_superior = (p + margem_erro) * 100

    # EXIBIÇÃO DOS RESULTADOS
    print("-" * 60)
    print(f"PAINEL DE MÉTRICAS DA ANÁLISE")
    print(f"VALOR ALVO: {valor_alvo.upper()}")
    print(f"Meta de 95% de confiança e 5% de erro na fase de planejamento")
    print("-" * 60)
    print(f"Tamanho total da amostra (n): {n}")
    print(f"Votos absolutos na amostra:   {votos_candidato}")
    print(f"Proporção Amostral (p):       {p*100:.1f}%")
    print(f"Margem de Erro amostra:       {margem_erro*100:.1f}%")
    print("-" * 60)

    print("\nRELATÓRIO TÉCNICO OFICIAL:")
    print("=" * 60)
    print(
        f"Com base na coleta de dados não probabilistica realizada  com {n}\n"
        f"indivíduos, foi identificada uma proporção para o valor alvo ({valor_alvo})\n"
        f"no valor de {p*100:.1f}%."
    )
    print("=" * 60)
```

```python
# Técnica de Amostragem (Probabilística - Estratificada)

# Definir estratos (1: com diabetes / 0:sem diabetes)
diabetes = df['diabetes'].value_counts(normalize=True)*100
print(diabetes)

n_sem_diabetes = int(n * 0.915)
n_com_diabetes = int(n * 0.085)
```

```python
# Coleta de dados
estrato_0 = df[df['diabetes'] == 0].sample(n=n_sem_diabetes)
estrato_1 = df[df['diabetes'] == 1].sample(n=n_com_diabetes)

dados_amostrais = pd.concat([estrato_0, estrato_1])
dados_amostrais.head(171)
```

```python
# Analisar dados da amostra

estimativa_pontual = np.mean(dados_amostrais['hemoglobina_glicada']) # média
variabilidade = np.std(dados_amostrais['hemoglobina_glicada'])       # desvpad

print(f"Estimativa pontual: {estimativa_pontual:.1f}%")
print(f"Variabilidade: {variabilidade:.1f}%")
print(f"Min: {np.min(dados_amostrais['hemoglobina_glicada'])}%")
print(f"Max: {np.max(dados_amostrais['hemoglobina_glicada'])}%")
```

```python
# Análise do erro amostral

# 1. Erro Padrão (EP)
EP = variabilidade / np.sqrt(len(dados_amostrais))
print(f"EP: {EP:.2f}%")

# 2. Margem Erro Final (E_final)
E_final = Z * EP
print(f"E_final: {E_final:.2f}%")

# 3. Intervalo de Confiança
LI = estimativa_pontual - E_final
LS = estimativa_pontual + E_final
print(f"IC: de {LI:.1f}% até {LS:.1f}%")
```
