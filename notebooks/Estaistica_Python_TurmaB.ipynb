```python
######################################
#### Medidas de tendência central ####
######################################

# Média Simples
media_simples1 = idade.mean()
media_simples2 = np.mean(idade)

# Média ponderada
valores = [8.5, 8.0, 3.0]
pesos   = [0.15, 0.35, 0.50]
media_ponderada = np.average(valores, weights=pesos)

# Média Aparada
from scipy.stats import trim_mean
media_aparada = trim_mean(idade.dropna(), 0.15)

# Mediana
mediana1 = np.median(idade)
mediana2 = np.nanmedian(idade)

# Moda
import statistics
moda1 = statistics.mode(trabalho)
moda2 = trabalho.mode()

valor = [1, 1, 2, 3, 4, 4]
moda3 = statistics.multimode(valor)
moda4 = statistics.mode(origem)


# Imprimir resultados
print(f'Média aritmética ponderada: {media_ponderada:.2f}')
print(f'Média aritmética simples: {media_simples2:.2f}')
print(f'Média aritmética simples: {media_simples2:.2f}')
print(f'Média aritmética aparada: {media_aparada:.2f}')
print(f'Mediana: {mediana2:.2f}')
print(f'Moda unimodal: {moda1}')
print(f'Moda multimodal: {moda3}')
print(f'Moda cidade origem: {moda4}')
```

## **Medidas de dispersão**

```python
# Importar as bibliotecas
import pandas as pd
import numpy as np
```

```python
# Carrergar os dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmB.xlsx",
                   sheet_name="DadosTratados2")

df.info()
```

```python
# Selecionar as variáveis que serão analisadas
leitura_bnu = df[df['C_origem'] == 'Blumenau']['Leituras'].tolist()
print(leitura_bnu)

leitura_outros = df[df['C_origem'] != 'Blumenau']['Leituras'].tolist()
print(leitura_outros)
```

```python
# Medidas de Posição
media_bnu    = np.mean(leitura_bnu)
media_outros = np.mean(leitura_outros)

mediana_bnu    = np.median(leitura_bnu)
mediana_outros = np.median(leitura_outros)

# Calcular medidas de dispersão

### Amplitude total
amp_bnu = max(leitura_bnu) - min(leitura_bnu)
amp_out = max(leitura_outros) - min(leitura_outros)

### Desvio Médio
mae_bnu = np.mean(np.abs(leitura_bnu - media_bnu))
mae_out = np.mean(np.abs(leitura_outros - media_outros))

### Variância
var_bnu = np.var(leitura_bnu, ddof=1)
var_out = np.var(leitura_outros, ddof=1)

### Desvio Padrão
std_bnu = np.std(leitura_bnu, ddof=1)
std_out = np.std(leitura_outros, ddof=1)

### Coeficiente de Variação
cv_bnu = (std_bnu / media_bnu)*100
cv_out = (std_out / media_outros)*100

# Imprimir resultados
print(f'Blumenau: Média = {media_bnu:.2f}, Mediana = {mediana_bnu:.2f}')
print(f'Outros: Média = {media_outros:.2f}, Mediana = {mediana_outros:.2f}\n')
print(f'Blumenau: Amplitude = {amp_bnu:.2f}')
print(f'Outros: Amplitude = {amp_out:.2f}\n')
print(f'Blumenau: MAE = {mae_bnu:.2f}')
print(f'Outros: MAE = {mae_out:.2f}\n')
print(f'Blumenau: Variância = {var_bnu:.2f}')
print(f'Outros: Variância = {var_out:.2f}\n')
print(f'Blumenau: Desvio Padrão = {std_bnu:.2f}')
print(f'Outros: Desvio Padrão = {std_out:.2f}\n')
print(f'Blumenau: Coeficiente de Variação = {cv_bnu:.2f}%')
print(f'Outros: Coeficiente de Variação = {cv_out:.2f}%')
```

# **Definição de Funções**

```python
# Importar as bibliotecas
import pandas as pd
import numpy as np
import statistics
```

```python
# Carrergar os dados

df = pd.read_excel("[Estatistica-BCD]_DadosTratados_CalourosBCD_turmB.xlsx",
                   sheet_name="DadosTratados2")
```

```python
def descritiva_quanti(df, variavel):

  # df = DataFrame com dados
  # variável = Nome da coluna analisada

  #1. Filtrar a variável do DataFrame
  coluna = df[variavel]

  #2. Calcular as medidas de posição
  media = round(np.mean(coluna), 2)
  mediana = round(np.median(coluna), 2)

  #3. Calcular as medidas de dispersão
  maximo = round(max(coluna), 2)
  minimo = round(min(coluna), 2)
  amplitude = round(maximo - minimo, 2)
  desvpad = round(np.std(coluna, ddof=1), 2)
  cv = round((desvpad / media)*100, 2)

  #4. Criar um dicionário com os resultados
  resultados = {
      "Variável":variavel,
      'Média': media,
      'Mediana': mediana,
      'Máximo': maximo,
      'Mínimo': minimo,
      'Amplitude': amplitude,
      'Desvio Padrão': desvpad,
      'Coeficiente de Variação': cv}

  #5. Converter o diconário dos resultados em um DataFrame
  df_final = pd.DataFrame([resultados])

  #6. Definir o retorno (saída) da função
  return df_final
```

```python
def descritiva_quali(df, variavel):

  # df = DataFrame com dados
  # variável = Nome da coluna analisada (string)

  #1. Filtrar a variável do DataFrame
  coluna = df[variavel]

  #2. Calcular as medidas de posição
  moda = statistics.multimode(coluna)

  #3. Calcular as medidas de dispersão
  n   = len(coluna.dropna())        # Total de Valores não nulos
  f_m = coluna.value_counts().max() # Frequência Máxima
  # Cálculo da razão de Variância
  rv = round(1 - (f_m/n),2)

  #4. Criar um dicionário com os resultados
  resultados = {
      "Variável":variavel,
      'Moda': moda,
      'RV': rv}

  #5. Converter o diconário dos resultados em um DataFrame
  df_final = pd.DataFrame([resultados])

  #6. Definir o retorno (saída) da função
  return df_final
```

# **Assimetria e Curtose**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gerar 1000 números inteiros aleatórios entre 17 e 60
np.random.seed(42)   # Garante reprodutibilidade
lista = np.random.randint(low=17, high=60, size=1000)
valores = pd.Series(lista)

# Cálculo da Assimetria
assimetria = valores.skew()

# Cálculo da Curtose
curtose = valores.kurtosis()

print(f"Assimetria: {assimetria:.2f}")
print(f"Curtose: {curtose:.2f}")
```

```python
from scipy.stats import gaussian_kde

# Criando um histograma
plt.hist(valores, bins=11, density=True, color='seagreen', edgecolor='black', alpha=0.5)

# Calculando e desenhando a curva de Densidade (KDE)
kde = gaussian_kde(valores)
eixo_x = np.linspace(min(valores), max(valores), 200) # Cria pontos do início ao fim do gráfico
plt.plot(eixo_x, kde(eixo_x), color='darkgreen', linewidth=2, label='Densidade (KDE)')

# Adicionando as linhas de Média e Mediana (Exatamente igual ao seu)
plt.axvline(valores.mean(), color='red', linestyle='--', label=f'Média: {valores.mean():.2f}')
plt.axvline(valores.median(), color='blue', linestyle='-', label=f'Mediana: {valores.median():.2f}')

# Personalização
plt.title(f'Análise da Forma da Distribuição\nAssimetria: {assimetria:.2f} | Curtose: {curtose:.2f}',
          fontsize=14, fontweight='bold')
plt.xlabel('Valores', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend()

plt.tight_layout()

plt.show()
```

```python
"""
Estudo da Assimetria e Curtose
"""

# Gerando dados aleatórios com o Numpy
np.random.seed(42)
n = 1000
min = 17
max = 60

normal   = np.clip(np.random.normal(43, 8, n), min, max)
ass_pos  = np.clip(np.random.gamma(shape=2, scale=10, size=n) + 17, min, max)
ass_neg  = np.clip(max - np.random.gamma(shape=2, scale=10, size=n), min, max)
curt_pos = np.clip(np.random.laplace(43, 5, n), min, max)
curt_neg = np.random.uniform(min, max, n)

# Criando a estrutura da figura
fig, axes = plt.subplots(3, 2, figsize=(12, 15))
axes = axes.flatten() # Transforma a grade em uma lista: axes[0], axes[1]...

# --- GRÁFICO 1: NORMAL ---
ax = axes[0]
ax.hist(normal, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(normal).plot(kind='kde', ax=ax, color='seagreen')
ax.axvline(np.mean(normal), color='red', linestyle='--', label='Média')
ax.axvline(np.median(normal), color='blue', label='Mediana')
ax.set_title(f"Normal\nAssimetria: {pd.Series(normal).skew():.2f}")
ax.legend()

# --- GRÁFICO 2: ASSIMETRIA POSITIVA ---
ax = axes[1]
ax.hist(ass_pos, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(ass_pos).plot(kind='kde', ax=ax, color='seagreen')
ax.axvline(np.mean(ass_pos), color='red', linestyle='--')
ax.axvline(np.median(ass_pos), color='blue')
ax.set_title(f"Assimetria Positiva\nAssimetria: {pd.Series(ass_pos).skew():.2f}")

# --- GRÁFICO 3: ASSIMETRIA NEGATIVA ---
ax = axes[2]
ax.hist(ass_neg, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(ass_neg).plot(kind='kde', ax=ax, color='seagreen')
ax.axvline(np.mean(ass_neg), color='red', linestyle='--')
ax.axvline(np.median(ass_neg), color='blue')
ax.set_title(f"Assimetria Negativa\nAssimetria: {pd.Series(ass_neg).skew():.2f}")

# --- GRÁFICO 4: CURTOSE POSITIVA ---
ax = axes[3]
ax.hist(curt_pos, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(curt_pos).plot(kind='kde', ax=ax, color='seagreen')
ax.set_title(f"Curtose Positiva\nCurtose: {pd.Series(curt_pos).kurt():.2f}")

# --- GRÁFICO 5: CURTOSE NEGATIVA ---
ax = axes[4]
ax.hist(curt_neg, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(curt_neg).plot(kind='kde', ax=ax, color='seagreen')
ax.set_title(f"Curtose Negativa\nCurtose: {pd.Series(curt_neg).kurt():.2f}")

# --- GRÁFICO 6: CURTOSE NORMAL (MESOCÚRTICA) ---
ax = axes[5]
ax.hist(normal, bins=12, density=True, color='lightgray', edgecolor='white')
pd.Series(normal).plot(kind='kde', ax=ax, color='seagreen')
curt_normal_val = pd.Series(normal).kurt()
ax.set_title(f"Curtose Normal (Mesocúrtica)\nCurtose: {curt_normal_val:.2f}")

plt.tight_layout()
plt.show()
```

# **Amostragem**

```python
import pandas as pd
import numpy as np
```

```python
# 1. Criando uma População Fictícia (N = 500 alunos)
np.random.seed(42)
populacao = pd.DataFrame({
    'ID': range(1, 501),
    'Nota_Final': np.random.normal(7, 1.5, 500).clip(0, 10),
    'Presenca': np.random.randint(70, 100, 500)
    })
```

## _Amostragem Aleatórria Simples_

```python
# 2. Definindo o tamanho da amostra
tamanho_amostra = 23

# 3. Executando a Amostragem Aleatória Simples
amostra_simples = populacao.sample(n=tamanho_amostra, random_state=42)

# 4. Comparando as médias
print(f"Média da Nota na População (N=500): {populacao['Nota_Final'].mean():.2f}")
print(f"Média da Nota na Amostra (n=50): {amostra_simples['Nota_Final'].mean():.2f}")
```

## _Amostragem sistemática_

```python
# 2. Definição do intervalo do passo (k = N / n)
N = len(populacao)
n = 50      # Tamanho da amostra
k = N // n  # // retorna a parte interira da divisão (sem arredondamento)

# 3. Escolha do ponto de partida aleatório
# Deve ser um número entre 0 e o valor do salto (k-1)
ponto_partida = np.random.randint(0, k)

# 4. Seleção da sequência sistemática
# Criamos uma lista de índices: partida, partida + k, partida + 2k +...
indices_sistematicos = np.arange(ponto_partida, N, step=k)
amostra_sistematica = populacao.iloc[indices_sistematicos]

# 5. Comparando as médias
print(f"Ponto de partida sorteado: {ponto_partida}")
print(f"Intervalo de salto (k): {k}")
print(f"Média da Nota na População: {populacao['Nota_Final'].mean():.2f}")
print(f"Média da Nota na Amostra Sistemática: {amostra_sistematica['Nota_Final'].mean():.2f}")
```

## _Amostragem Estratificada_

```python
# 1. Criando a população com estratos de gênero
np.random.seed(42)
n_total = 500

# Vamos simular uma proporção de 50% Mulheres, 42% Homens e 8% Outros
populacao = pd.DataFrame({
    'ID': range(1, n_total + 1),
    'Genero': np.random.choice(['Feminino', 'Masculino', 'Outros'],
                                size=n_total, p=[0.50, 0.42, 0.08]),
    'Nota_Final': np.random.normal(7, 1.5, n_total).clip(0, 10).round(2)
    })

# 2. Executando a Amostragem Estratificada
# Vamos extrair uma amostra de 10% (n = 50 alunos)
amostra_estratificada = populacao.groupby('Genero', group_keys=False).apply(
    lambda x: x.sample(frac=0.10, random_state=42))

# 3. Comparação de proporcionalidade
print("--- Distribuição na POPULAÇÃO (Real) ---")
print(populacao['Genero'].value_counts(normalize=True).round(2))

print("\n--- Distribuição na AMOSTRA (Estratificada) ---")
print(amostra_estratificada['Genero'].value_counts(normalize=True).round(2))

# 4. Comparação das médias
media_pop = populacao['Nota_Final'].mean()
media_amostra = amostra_estratificada['Nota_Final'].mean()

print("\n" + "="*40)
print(f"Média da Nota (População): {media_pop:.2f}")
print(f"Média da Nota (Amostra):   {media_amostra:.2f}")
print("="*40)
```
